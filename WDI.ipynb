```python
import re

def to_concept_id(s):
    '''convert a string to lowercase alphanumeric + underscore id for concepts'''
    return re.sub(r'[/ -\.]+', '_', s).lower()
```

```python
def simple_df(s1, ser):
    da = ser[s1].dropna().unique()
    df = pd.DataFrame([], columns=[ to_concept_id(s1), 'name'])
    df['name'] = da
    df[to_concept_id(s1)] = df['name'].apply(to_concept_id)
#     df = df.reset_index()
    df.columns = [to_concept_id(s1), 'name']
    
    return df
```
